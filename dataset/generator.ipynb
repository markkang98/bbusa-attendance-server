```python
from datetime import date, timedelta, datetime
from dateutil.parser import parse
saturdays = []
sundays = []

def allsundays(year):
   d = datetime(year, 1, 1, 0, 0, 0)                    # January 1st
   d += timedelta(days = 6 - d.weekday())  # First Sunday
   while d.year == year:
      yield d
      d += timedelta(days = 7)
    
def allsaturdays(year):
   d = datetime(year, 1, 1, 0, 0, 0)                    # January 1st
   d += timedelta(days = 5 - d.weekday())  # First Sunday
   while d.year == year:
      yield d
      d += timedelta(days = 7)

for year in [2017, 2018, 2019]:
    sundays.extend(allsundays(year))
    saturdays.extend(allsaturdays(year))

takes = pd.read_csv('Takes.csv', delimiter = ',')

STC = pd.concat([students, takes, classes], axis=1, join='inner')
STC = STC.loc[:, ~STC.columns.duplicated()]
alldates = []
dates = []
record = []
attSID = []
attCID = []

for i, row in STC.iterrows():
    if (row['day'] == 'Sunday'):
        alldates = sundays
    elif (row['day'] == 'Saturday'):
        alldates = saturdays
    for date in alldates:
        startDate = datetime.strptime(row['startDate'], '%Y-%m-%d')
        if (date > startDate):
            attSID.append(int(row['SID']))
            attCID.append(int(row['CID']))
            dates.append(date)
            rand = random.randint(1, 100)
            if (rand <= 80):
                record.append('attended')
            elif (rand <= 90):
                record.append('excused')
            else:
                record.append('absent')


d = {'SID': attSID, 'CID': attCID, 'date': dates, 'attendance': record}
attendance = pd.DataFrame(data=d)
attendance.to_csv('Attendance.csv')
```
